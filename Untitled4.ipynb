```python
import pandas as pd
data = pd.read_csv("teams.csv")
print(data.head())
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("teams.csv")

# Comparer buts marqués vs buts encaissés
plt.figure(figsize=(10,6))
plt.scatter(df["goals"], df["goals_against"], color="red")
for i, team in enumerate(df["team"]):
    plt.text(df["goals"].iloc[i]+0.1, df["goals_against"].iloc[i]+0.1, team, fontsize=8)

plt.xlabel("Buts marqués (Offensif)")
plt.ylabel("Buts encaissés (Défensif)")
plt.title("Performance offensive vs défensive - FIFA World Cup 2026")
plt.show()
```

```python
# Calculer la corrélation entre les buts marqués et les buts encaissés
correlation = df['goals'].corr(df['goals_against'])
print(f"La corrélation entre les buts marqués et les buts encaissés est : {correlation:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le dataset
df = pd.read_csv("teams.csv")

# Aperçu
df.head()
plt.figure(figsize=(10,6))
plt.scatter(df["goals"], df["goals_against"], color="red")

# Ajouter les noms des équipes
for i, team in enumerate(df["team"]):
    plt.text(df["goals"].iloc[i]+0.1, df["goals_against"].iloc[i]+0.1, team, fontsize=8)

plt.xlabel("Buts marqués (Offensif)")
plt.ylabel("Buts encaissés (Défensif)")
plt.title("Performance offensive vs défensive - FIFA World Cup 2026")
plt.show()
df["performance_index"] = df["goals"] - df["goals_against"]

df.sort_values("performance_index", ascending=False, inplace=True)

plt.figure(figsize=(12,6))
plt.barh(df["team"], df["performance_index"], color="skyblue")
plt.xlabel("Indice de performance (buts marqués - buts encaissés)")
plt.title("Indice de performance offensive vs défensive par équipe")
plt.show()
plt.figure(figsize=(12,8))
sns.heatmap(df[["goals","assists","shots_on_target_pct","goals_against","shots_against","gk_save_pct"]].corr(),
            annot=True, cmap="coolwarm")
plt.title("Corrélations entre variables offensives et défensives")
plt.show()

```
